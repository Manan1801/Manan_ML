Q.13 Write a function to compute the argmin of a numpy array. The function should take a numpy array as input and return the index of the minimum element. You can use the np.argmin function to verify your solution.

```python
arr = np.random.randint(0,10,6)
def mini(ls):
    m=np.inf
    a=0
    for i in range(len(ls)):
        if ls[i]<m:
            m=ls[i]
            a=i
    return a
print(arr)
print(mini(arr))
print(np.argmin(arr))
```

Q.6 Use sympy to confirm that you obtain the same gradient analytically.


```python
from sympy import symbols, diff, sin

a, b = symbols('a b')


print(a, b)


function = ((a**2) * b) + ((b**3) * sin(a))


diff_a = diff(function, a)
diff_b = diff(function, b)


values = {a: 3, b: 6}


d_a = diff_a.subs(values)
d_b = diff_b.subs(values)


print(float(d_a))
print(float(d_b))
```

Q.7 Create a Python nested dictionary to represent hierarchical information. We want to store record of students and their marks. Something like:


```python
# Creating the nested dictionary directly
university = {
    '2021': {
        'Computer Science': [
            {'Roll Number': 101, 'Name': 'Alice', 'Scores': {'Math': 90, 'Physics': 85}},
            {'Roll Number': 102, 'Name': 'Bob', 'Scores': {'Math': 80, 'Physics': 88}}
        ],
        'Mechanical': [
            {'Roll Number': 103, 'Name': 'Charlie', 'Scores': {'Math': 75, 'Physics': 80}}
        ]
    },
    '2022': {
        'Electrical': [
            {'Roll Number': 104, 'Name': 'David', 'Scores': {'Math': 85, 'Physics': 87}},
            {'Roll Number': 105, 'Name': 'Eve', 'Scores': {'Math': 78, 'Physics': 82}}
        ],
        'Civil': [
            {'Roll Number': 106, 'Name': 'Frank', 'Scores': {'Math': 82, 'Physics': 79}}
        ]
    },
    '2023': {
        'Computer Science': [
            {'Roll Number': 107, 'Name': 'Grace', 'Scores': {'Math': 89, 'Physics': 84}},
            {'Roll Number': 108, 'Name': 'Hank', 'Scores': {'Math': 81, 'Physics': 86}}
        ],
        'Mechanical': [
            {'Roll Number': 109, 'Name': 'Ivy', 'Scores': {'Math': 76, 'Physics': 83}},
            {'Roll Number': 110, 'Name': 'Jack', 'Scores': {'Math': 74, 'Physics': 80}}
        ]
    }
}

# Printing the university data
for year, departments in university.items():
    print(f"Year: {year}")
    for department, students in departments.items():
        print(f"  Department: {department}")
        for student in students:
            print(f"    Roll Number: {student['Roll Number']}, Name: {student['Name']}, Scores: {student['Scores']}")
```

```python
import jax.numpy as jnp
from jax import grad
```

```python
def sum_logistic(x,y):
  return ((x**2)*y) + ((y**3)*jnp.sin(x))
x=3.0
y=6.0
derivative_fn_x = grad(sum_logistic,argnums=0)
derivative_fn_y = grad(sum_logistic,argnums=1)
print(derivative_fn_x(x,y))
print(derivative_fn_y(x,y))
```
